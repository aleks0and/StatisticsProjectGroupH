```python
#importing libraries and loading dataset

import numpy as np
import pandas as pd

weather_data = pd.read_csv(r"C:\Users\Michal\Desktop\School\Bocconi\Courses\Business Analytics\ProbabilityAndStatistics-master\ProbabilityAndStatistics-master\Assignment_1\data\nycflights13\nycflights13_weather.csv", 
                           sep = ",", 
                           names=["origin","year","month","day","hour","temp","dewp","humid","wind_dir","wind_speed","wind_gust","precip","pressure","visib","time_hour"], 
                           skiprows = [0])
```

```python
weather_data.head(10)
```

```python
#Converting temperatures into celsius

def tempConverter(temperature):
    temperature = (temperature - 32) * (5/9)
    return float(temperature)

weather_data["temp"] = weather_data["temp"].apply(tempConverter)
```

```python
weather_data.head(20)
```

```python
#Computing daily mean temperatures for JFK

weather_data_JFK = weather_data[weather_data["origin"] == "JFK"].interpolate().groupby("day").mean()
weather_data_JFK["temp"].head(20)

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('Days')
plt.ylabel('Daily mean temperatures')
plt.title("Daily mean temperatures for January 2013 at JFK Airport")
plt.plot(weather_data_JFK["temp"])
```

```python
#Days with greater mean temperature than in preceding day

better_days = weather_data_JFK[(weather_data_JFK.shift(periods=1)["temp"] - weather_data_JFK["temp"]) > 0]

better_days
```

```python
#5 warmest days

best_days = weather_data_JFK.sort_values(by = ["temp"], ascending = False).head(5)
```
