```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.decomposition import PCA 

data = pd.read_csv("wines_properties.csv")
```

```python
data.head()
```

```python
#num 1 
x = data.iloc[:,1:]
y = data["Alcohol"]
#sm.OLS doesn't automatically add a constant
x = sm.add_constant(x)

#ols model
mod = sm.OLS(y, x)
res = mod.fit()

print(res.summary())
```

```python
#num 2 
treshold = np.percentile(data["Alcohol"], 75)

#create dummy
data["strong"] = 1
data.loc[data["Alcohol"] < treshold, "strong"] = 0 #put the right threshold here instead of 13

#i don't this is not necessary
#dummy =  pd.get_dummies(data["strong"], prefix = "strong")
#data_complete = data.join(dummy.loc[:,:"strong_0"])
#data_complete.head()

data.head()
#setting the model
logistical_regression = sm.Logit(data["strong"],x)

#fitting the model
fitted_model = logistical_regression.fit()
fitted_model.summary2()
```
